```python
import pandas as pd
import glob
```

```python
refs = pd.read_csv('..\\..\\refs\\refs.csv')
links = pd.DataFrame()
```

```python
csv_files = glob.glob('../../Sefaria-Export/links/links*.csv')
for csv_file in csv_files:
    df = pd.read_csv(csv_file)
    links = pd.concat([links, df])
    
```

add a reverse copy of the links and concat the two dataframes

```python
reversed_links = pd.DataFrame()
reversed_links['Citation 1'] = links['Citation 2']
reversed_links['Citation 2'] = links['Citation 1']
reversed_links['Conection Type'] = links['Conection Type']
links = pd.concat([links, reversed_links])
```

```python
merged_df_citation1 = links.merge(refs, left_on='Citation 1', right_on='ref', how='left')
```

```python
merged_df_citation1['path'].unique().__len__()
```

```python
merged_df_citation1 = links.merge(refs, left_on='Citation 1', right_on='ref', how='left')
merged_df = merged_df_citation1.merge(refs, left_on='Citation 2', right_on='ref', how='left', suffixes=('_1', '_2'))
```

```python
merged_df.head()
```

```python
result_df = merged_df.dropna(subset=['path_1', 'path_2'])
```

```python
result_df
```

```python
result_df = result_df[['path_1','line_index_1','heRef_2','path_2', 'line_index_2','Conection Type']]
```

```python
unique_paths = result_df['path_1'].unique()

for path in unique_paths:
    path_df = result_df[result_df['path_1'] == path]
    path_df.drop(['path_1'], axis=1, inplace=True)
    path = path.split('\\')[-1]
    path_df.to_json(f'../links/{path}_links.json', orient='records')
```
